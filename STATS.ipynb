**Statistics Basics**

1 What is statistics, and why is it importanT?

ans Statistics is the science of collecting, analyzing, interpreting, and presenting data. It helps in making informed decisions, identifying trends, testing hypotheses, and drawing conclusions from data in fields like business, medicine, and social sciences.

2 What are the two main types of statistics?

ANS Descriptive Statistics : Summarizes and describes data (e.g., mean, median, graphs).

Inferential Statistics : Draws conclusions about a population from sample data (e.g., hypothesis testing, confidence intervals).

3 What are descriptive statistics?

ANS Methods to summarize and describe key features of data (e.g., measures of central tendency, dispersion, and visualizations like histograms).

4 What is inferential statistics?

ANS Techniques to make predictions or generalizations about a population based on sample data (e.g., regression, t-tests)

5 What is sampling in statistics?

ANS The process of selecting a subset (sample) from a population to estimate characteristics of the whole group.

6 What are the different types of sampling methods?

ANS Random: Simple random, stratified, cluster, systematic.

Non-Random: Convenience, purposive, quota, snowball.

7 What is the difference between random and non-random sampling?

ANS Random: Every member has an equal chance of selection (unbiased).

Non-Random: Selection based on convenience or judgment (may introduce bias)

8 Define and give examples of qualitative and quantitative data?

ANS Qualitative: Non-numerical (e.g., gender, opinions).

Quantitative: Numerical (e.g., height, temperature).

9 What are the different types of data in statistics?

ANS Qualitative data: Non-numerical (e.g., gender, opinions).

a. Discreet data: finite numbers(whole numbers)

b. Continuous data: any value in a range(movie duration)

Quantitative data: Numerical (e.g., height, temperature).

a. Nominal data: without any order(gender)

b. Ordinal data: with order(rank: good, better, best)

10 Explain nominal, ordinal, interval, and ratio levels of measurement?

ANS Nominal: Categories (e.g., colors).

Ordinal: Ordered ranks (e.g., survey scales).

Interval: Equal gaps, no true zero (e.g., Celsius).

Ratio: True zero (e.g., weight, height).

11 What is the measure of central tendency?

ANS A single value representing the center of a dataset (mean, median, mode).

12 Define mean, median, and mode?

ans Mean: Average (sum of values / count).

Median: Middle value in ordered data.

Mode: Most frequent value.

13 What is the significance of the measure of central tendency?

ans Provides a summary of the dataset’s "typical" value, aiding in comparisons and analysis.


14 What is variance, and how is it calculated?

ans Measures data spread. Calculated as the average of squared deviations from the mean.

Formula: s^2=∑(xi-x)^2(1/n-1)



15 What is standard deviation, and why is it important?

ans Square root of variance. Indicates how much data deviates from the mean. Importance: Easier interpretation (same units as data).

16 Define and explain the term range in statistics?

ans Difference between the highest and lowest values in a dataset.

17 What is the difference between variance and standard deviation?

ans Variance is in squared units.

SD is in the original units (more interpretable).

18 What is skewness in a dataset?

ans Measures asymmetry in data distribution:

Positive skew: Right tail longer.

Negative skew: Left tail longer.

19 What does it mean if a dataset is positively or negatively skewed?

ANS Positive Skew (Right-Skewed):

Tail extends to the right.

Mean > Median > Mode.

Example: Income distribution (few extremely high incomes).

Negative Skew (Left-Skewed):

Tail extends to the left.

Mean < Median < Mode.

Example: Exam scores (most students score high, few score low).

20 Define and explain kurtosis?

ANS Definition: Measures the "tailedness" of a distribution (peakedness + outlier risk).

Types:

Leptokurtic (High Kurtosis): Heavy tails, more outliers (e.g., financial returns).

Platykurtic (Low Kurtosis): Light tails, fewer outliers (e.g., uniform distribution).

Mesokurtic: Normal tails (e.g., Gaussian distribution).

21 What is the purpose of covariance?
ANS Measures how two variables vary together.

Positive Covariance: Both increase/decrease together.

Negative Covariance: One increases while the other decreases.

Limitation: Sensitive to units (hard to interpret magnitude).

24 What does correlation measure in statistics?

ANS Quantifies the strength and direction of a linear relationship between two variables.

Range: -1 (perfect negative) to +1 (perfect positive).

Example: Height vs. Weight (likely positive correlation).

23 What is the difference between covariance and correlation?

ANS Covariance measures how two variables change together (joint variability).

Correlation measures the strength and direction of their linear relationship.

24 What are some real-world applications of statistics?

ANS Healthcare: Drug efficacy testing, disease spread modeling.

Finance: Risk assessment, stock market analysis.

Marketing: Customer segmentation, A/B testing.

Sports: Player performance analytics.

Quality Control: Manufacturing defect detection (e.g., Six Sigma).

Weather Forecasting: Predictive modeling.

```python
##################################################
# 1. Calculate mean, median, and mode
##################################################
import statistics

data1 = [12, 15, 18, 22, 25, 25, 28, 30, 32, 35]

# Mean
mean = statistics.mean(data1)
print(f"Mean: {mean}")

# Median
median = statistics.median(data1)
print(f"Median: {median}")

# Mode
try:
    mode = statistics.mode(data1)
    print(f"Mode: {mode}")
except statistics.StatisticsError:
    print("No unique mode found")

##################################################
# 2. Variance and standard deviation
##################################################
import numpy as np

data2 = [12, 15, 17, 20, 22, 25, 28, 30, 32, 35]

# Variance (sample)
variance = np.var(data2, ddof=1)
print(f"Variance: {variance:.2f}")

# Standard Deviation (sample)
std_dev = np.std(data2, ddof=1)
print(f"Standard Deviation: {std_dev:.2f}")

##################################################
# 3. Data types: nominal, ordinal, interval, ratio
##################################################
# Nominal data
colors = ["red", "blue", "green", "blue", "red"]

# Ordinal data
education_levels = ["High School", "Bachelor's", "Master's", "PhD"]

# Interval data
temperatures_c = [20.5, 22.3, 18.7, 25.0]

# Ratio data
heights_cm = [165, 180, 172, 190, 168]

print("Nominal example:", colors[:3])
print("Ordinal example:", education_levels)
print("Interval example:", temperatures_c)
print("Ratio example:", heights_cm)

##################################################
# 4. Sampling: random and stratified
##################################################
import pandas as pd

np.random.seed(42)
df = pd.DataFrame({
    'age': np.random.randint(18, 65, 100),
    'income': np.random.randint(20000, 100000, 100)
})

# Simple Random Sampling
random_sample = df.sample(n=10)
print("\nRandom Sample (10 records):")
print(random_sample.head())

# Stratified Sampling by age group
df['age_group'] = pd.cut(df['age'], bins=[18, 30, 45, 65])
stratified_sample = df.groupby('age_group').apply(lambda x: x.sample(frac=0.2))
print("\nStratified Sample:")
print(stratified_sample.head())

##################################################
# 5. Calculate range of a dataset
##################################################
def calculate_range(dataset):
    return max(dataset) - min(dataset)

range_data = [45, 32, 56, 71, 63, 82, 47]
print(f"\nRange of data: {calculate_range(range_data)}")

##################################################
# 6. Histogram to visualize skewness
##################################################
import matplotlib.pyplot as plt
import seaborn as sns

right_skewed = np.random.exponential(scale=2, size=1000)
left_skewed = np.random.beta(a=5, b=2, size=1000) * 100

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(right_skewed, kde=True)
plt.title("Right (Positive) Skew")

plt.subplot(1, 2, 2)
sns.histplot(left_skewed, kde=True)
plt.title("Left (Negative) Skew")
plt.show()

##################################################
# 7. Calculate skewness and kurtosis
##################################################
from scipy.stats import skew, kurtosis

normal_data = np.random.normal(0, 1, 1000)
print(f"\nSkewness: {skew(normal_data):.2f}")
print(f"Kurtosis: {kurtosis(normal_data):.2f}")

##################################################
# 8. Demonstrate positive and negative skewness
##################################################
positive_skew = np.random.chisquare(df=3, size=1000)
negative_skew = 100 - np.random.chisquare(df=3, size=1000)

print("\nPositive skew mean vs median:")
print(f"Mean: {np.mean(positive_skew):.2f}, Median: {np.median(positive_skew):.2f}")

print("\nNegative skew mean vs median:")
print(f"Mean: {np.mean(negative_skew):.2f}, Median: {np.median(negative_skew):.2f}")

##################################################
# 9. Calculate covariance
##################################################
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 3, 4, 5, 6, 7, 8, 9, 10]

covariance = np.cov(x, y, ddof=1)[0, 1]
print(f"\nCovariance between x and y: {covariance:.2f}")

##################################################
# 10. Calculate correlation coefficient
##################################################
correlation = np.corrcoef(x, y)[0, 1]
print(f"\nPearson correlation coefficient: {correlation:.2f}")

##################################################
# 11. Scatter plot for relationship
##################################################
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.7)
plt.title("Scatter Plot of X vs Y")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.grid(True)
plt.show()

##################################################
# 12. Random vs systematic sampling
##################################################
def systematic_sample(dataframe, step):
    return dataframe.iloc[::step]

print("\nSimple Random Sample:")
print(df.sample(n=5))

print("\nSystematic Sample (every 10th record):")
print(systematic_sample(df, 10).head())

##################################################
# 13. Grouped data: mean and median
##################################################
bins = [0, 20, 40, 60, 80]
grouped = pd.cut(df['age'], bins=bins)

group_stats = df.groupby(grouped)['age'].agg(['mean', 'median'])
print("\nGrouped Data Statistics:")
print(group_stats)

##################################################
# 14. Simulate data and analyze
##################################################
sim_data = np.random.normal(50, 10, 1000)

print("\nSimulated Data Analysis:")
print(f"Mean: {np.mean(sim_data):.2f}")
print(f"Median: {np.median(sim_data):.2f}")
print(f"Standard Deviation: {np.std(sim_data, ddof=1):.2f}")
print(f"Variance: {np.var(sim_data, ddof=1):.2f}")
print(f"Range: {np.ptp(sim_data):.2f}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import iqr, zscore, pearsonr, spearmanr, skew, kurtosis

# Generate synthetic dataset
np.random.seed(42)
df = pd.DataFrame({
    'math': np.random.normal(70, 10, 100),
    'science': np.random.normal(65, 12, 100),
    'english': np.random.normal(75, 8, 100),
    'history': np.random.normal(60, 15, 100)
})

##################################################
# 15. Use NumPy or pandas to summarize descriptive statistics
##################################################
print("15. Descriptive Statistics:\n", df.describe())

##################################################
# 16. Plot a boxplot to understand the spread and identify outliers
##################################################
plt.figure(figsize=(8, 5))
sns.boxplot(data=df)
plt.title("16. Boxplot of Subjects")
plt.show()

##################################################
# 17. Calculate the Interquartile Range (IQR)
##################################################
iqr_math = iqr(df['math'])
print(f"17. IQR of Math scores: {iqr_math:.2f}")

##################################################
# 18. Implement Z-score normalization and explain its significance
##################################################
z_scores = df.apply(zscore)
print("\n18. Z-score Normalized Data (first 5 rows):\n", z_scores.head())
print("Significance: Z-score normalization transforms data to have mean=0 and std=1, useful for comparison across scales.")

##################################################
# 19. Compare two datasets using their standard deviations
##################################################
std_math = df['math'].std()
std_science = df['science'].std()
print(f"\n19. Standard Deviation - Math: {std_math:.2f}, Science: {std_science:.2f}")
if std_math > std_science:
    print("Math scores are more dispersed.")
else:
    print("Science scores are more dispersed.")

##################################################
# 20. Visualize covariance using a heatmap
##################################################
cov_matrix = df.cov()
plt.figure(figsize=(6, 4))
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm')
plt.title("20. Covariance Matrix Heatmap")
plt.show()

##################################################
# 21. Create a correlation matrix using seaborn
##################################################
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
plt.title("21. Correlation Matrix (Pearson)")
plt.show()

##################################################
# 22. Compute variance and standard deviation
##################################################
print("\n22. Variance and Standard Deviation:")
for column in df.columns:
    var = df[column].var()
    std = df[column].std()
    print(f"{column}: Variance = {var:.2f}, Std Dev = {std:.2f}")

##################################################
# 23. Visualize skewness and kurtosis
##################################################
plt.figure(figsize=(12, 5))
for i, column in enumerate(df.columns, 1):
    plt.subplot(1, 4, i)
    sns.histplot(df[column], kde=True)
    plt.title(f"{column}\nSkew: {skew(df[column]):.2f}, Kurt: {kurtosis(df[column]):.2f}")
plt.tight_layout()
plt.show()

##################################################
# 24. Implement Pearson and Spearman correlation coefficients
##################################################
pearson_corr, _ = pearsonr(df['math'], df['science'])
spearman_corr, _ = spearmanr(df['math'], df['science'])
print(f"\n24. Pearson Correlation (Math vs Science): {pearson_corr:.2f}")
print(f"Spearman Correlation (Math vs Science): {spearman_corr:.2f}")
```
